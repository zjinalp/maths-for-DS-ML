```python
import numpy as np
import pandas as pd
```

##### Find outlier in terms of revenue

```python
df = pd.read_csv("movie_revenues.csv")
df.head(3)
```

```python
#  revenue in million
df["revenue_in_million"] = df["revenue"].apply(lambda x : x/1000000)
df["revenue_in_million"].describe()
```

```python
df["revenue_in_million"].head()
```

```python
# /find mean, std 
_, mean, std, *_ = df["revenue_in_million"].describe()
mean, std
```

```python
# for z score
def get_z_score(value, mean, std) :
    return (value - mean)/std
```

```python
# make column for z score
df["z-score"] =  df["revenue_in_million"].apply(lambda x : get_z_score(x, mean, std))
df["z-score"].head()
```

```python
df[df["z-score"] > 3]
```

```python
#  for modified z - score

def get_mad(s): # here s is series
    median = np.median(s)
    diff = abs(s-median)
    MAD = np.median(diff)
    return MAD
```

```python
MAD = get_mad(df.revenue_in_million)
median = np.median(df.revenue_in_million)
MAD, median
```

```python
def get_modified_z_score(x, median, MAD):
    return 0.6745*(x-median)/MAD
```

```python
get_modified_z_score(2787, median, MAD)
```

```python
df['mod_z_score'] = df.revenue_in_million.apply(lambda x: get_modified_z_score(x, median, MAD))
df.head(3)
```

```python
df[df.mod_z_score>3.5]
```
